```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 1. 삼성전자 데이터 가져오기
df = yf.download("005930.KS", start="2025-01-01")

# 2. 20일 이동평균선과 60일 이동평균선 계산
df['MA20'] = df['Close'].rolling(window=20).mean()
df['MA60'] = df['Close'].rolling(window=60).mean()

# 3. 주가와 이동평균선을 함께 시각화
plt.figure(figsize=(14, 7))
plt.plot(df['Close'], label='Close Price', color='gray', alpha=0.5)
plt.plot(df['MA20'], label='20-day MA', color='red')
plt.plot(df['MA60'], label='60-day MA', color='blue')

plt.title('Samsung Stock Moving Average')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')  # 모든 경고 메시지 숨기기
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # 예쁜 시각화를 위한 라이브러리

# 1. 반도체 3사 주가 한 번에 가져오기
tickers = ['NVDA', 'TSM', '005930.KS']
data = yf.download(tickers, start="2025-01-01")['Close']

# 2. 매일매일의 수익률 변동 계산하기
returns = data.pct_change()

# 3. 종목 간의 상관계수 구하기 (1에 가까울수록 똑같이 움직임)
corr = returns.corr()

# 4. 히트맵(Heatmap)으로 시각화하기
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Semiconductor Stocks Correlation')
plt.show()
```
